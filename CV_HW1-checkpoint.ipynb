# Загружаем библиотеки

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

### Загружаем картинку

```python
img = cv2.imread("me.jpg")
```

Размер изображения

```python
img.shape
```

```python
plt.imshow(img)
```

Явное преобразование

```python
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

### 1. Найдём лицо на изображении, используя каскадный классификатор

```python
haar_face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
face = haar_face_cascade.detectMultiScale(img)
print('Face found: ', len(face))
```

Выделим лицо на изображении

```python
imgCopy = img.copy()
for (x, y, w, h) in face:
    cv2.rectangle(imgCopy, (x, y), (x+w, y+h), (0, 255, 0), 1)
plt.imshow(imgCopy, cmap="gray")
```

### 2. Получим фрагмент изображения, используя срезы

```python
for (x, y, w, h) in face:
    cropped = img[y:y+h, x:x+w]
plt.imshow(cropped, cmap="gray")
```

Отступим на 10% от границ лица

```python
h, w
```

```python
h*0.1, w*0.1
```

```python
int(h*0.1)
```

```python
for (x, y, w, h) in face:
    deltaY=int((h*0.1)/2)
    deltaX=int((w*0.1)/2)
    cropped = img[y-deltaY:y+h+deltaY, x-deltaX:x+w+deltaX]
plt.imshow(cropped, cmap="gray")
```

### 3. Получим бинарное изображение краев (границ объекта)

```python
img2 = cv2.cvtColor(cropped, cv2.COLOR_BGR2GRAY)
```

```python
edges = cv2.Canny(img2, 50, 150)
plt.imshow(edges, cmap="gray")
```

### 4. Удалим мелкие границы, у которых длина и ширина меньше 10

Если удалять через площадь:

```python
edges1=edges.copy()
contours, _ = cv2.findContours(edges1, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
#contours = sorted(contours, key = cv2.contourArea, reverse = True)
for contour in contours:
    if (cv2.contourArea(contour) < 9):
        cv2.drawContours(edges1, contour, -1, (0, 0, 0), 1)
plt.imshow(edges1, cmap="gray")
```

Удалим все связные компоненты, у которых длина и ширина меньше 10

```python
w, h = edges.shape
_, labels, stats, _ = cv2.connectedComponentsWithStats(edges, 4)
for i in range(w):
    for j in range(h):
        if stats[labels[i, j], cv2.CC_STAT_WIDTH] <= 10 and stats[labels[i, j], cv2.CC_STAT_HEIGHT] <= 10:
            edges[i, j] = 0
plt.imshow(edges, cmap="gray")
```

### 5. Применим морфологическую операцию наращивания (размер структурирующего элемента 5х5)

```python
kernel = np.ones((5,5), np.uint8) 
imgDilate = cv2.dilate(edges, kernel)
plt.imshow(imgDilate, cmap="gray")
```

### 6. Сгладим полученное изображение краёв гауссовским фильтром 5 на 5. 

```python
imgGauss=cv2.GaussianBlur(imgDilate,(5,5), cv2.BORDER_DEFAULT)
plt.imshow(imgGauss, cmap="gray")
```

### Получим нормализованное изображение М, где все пикселы от 0 до 1.

```python
imageNorm = cv2.normalize(imgGauss,None,0,1,cv2.NORM_MINMAX)
plt.imshow(imageNorm, cmap="gray")
```

### 7. Получим изображение F1 лица с примененной билатеральной фильтрацией.

```python
F1=cv2.bilateralFilter(cropped,3,75,75)
plt.imshow(F1, cmap="gray")
```

Медианная фильтрация

```python
Fmed=cv2.medianBlur(cropped,3)
plt.imshow(Fmed, cmap="gray")
```

Блюр

```python
Fblur=cv2.blur(cropped, (3,3))
plt.imshow(Fblur, cmap="gray")
```

### 8. Получим изображение F2 лица с улучшенной четкостью/контрастностью

```python
support =cv2.GaussianBlur(cropped, (0, 0), 2.0)
F2 = cv2.addWeighted(cropped, 2.0, support, -1.0, 0.0)
plt.imshow(F2, cmap="gray")
```

Как выглядело изображение до применения фильтра:

```python
plt.imshow(cropped, cmap="gray")
```

### 9. Осуществим финальную фильтрацию по формуле:

$$ 
Result[x,y,c] = M[x,y]*F2[x,y,c] + (1-M[x,y])*F1[x,y,c],
$$ 
$$
где x=1..width, y=1..height, c=1..3
$$

```python
result = np.zeros(cropped.shape, dtype=np.uint8)
w, h, c = result.shape
for i in range(w):
    for j in range(h):
        for k in range(c):
            result[i, j, k] = imageNorm[i, j] * F2[i, j, k] + (1 - imageNorm[i, j]) * F1[i, j, k]
plt.imshow(result, cmap="gray")
```

Как выглядело изображение до применения всех фильтров:

```python
plt.imshow(cropped, cmap="gray")
```
